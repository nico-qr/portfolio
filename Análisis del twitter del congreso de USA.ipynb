```python
import networkx as nx

G = nx.read_edgelist("congress.edgelist")
print(G)
```

```python
d = dict(G.degree)
nx.draw(G, nodelist = d.keys(), node_size = [v for v in d.values()])
```

```python
m = max(d.values())

def get(val):
    for (key, value) in d.items():
        if val == value:
            return key
        
    return None
        
print(get(m))
```
