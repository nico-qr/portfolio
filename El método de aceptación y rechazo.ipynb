En este notebook desarrollamos el algortimo de aceptación y rechazo para realizar un muestreo de una función de distribución de probabilidad (en inglés _**P**robability **D**istribution **F**unction_). 

## Marco teórico

### Motivación

Recordemos que dada una PDF puede interesarnos obtener un valor a partir de ella. Es decir: dada una función $f$ tomar un número de manera aleatoria siguiendo la distribución de la función. Sin embargo, en ocasiones no es posible obtener directamente el valor a partir de la función. 

Supongamos realizamos un estudio de satisfacción al cliente en el que cada persona puede otorgarnos una calificación (_score_) $s\in (-\infty, \infty)$. Nos interesa ahora  saber cuál es la probabilidad de tener cierta calificación, que denotaremos por $p(s)$. Este algoritmo nos ayuda a encontrar esa evaluación cuando no conocemos directamente la función $p$. 

Supongamos adicionalmente que conocemos una función $f$ tal que para cada $s$ se verifica que:

$$
p(s) = \frac{f(s)}{k}
$$

donde $k$ es una constante de normalización. Dado que en la práctica puede ser difícil encontrar esa constante, basta que trabajemos con $f$ para encontrar la evaluación de nuestro interés.

### Algoritmo

Supongamos entonces que conocemos la función $f$ que es más sencilla de computar que $p$. Ahora tomaremos una función de distribución arbitraria $g$ (potencialmente más sencilla de computar que $f$) tal que alguna $M$ cumpla que para cada $s$:

$$
f(s) \leq M g(s)
$$

Hecho esto procedemos con el siguiente algoritmo:

1. Tomamos un valor $x$ siguiendo la distribución $g$
2. Tomamos un valor $u$ en el intervalo [0,1] de manera uniforme
3. Aceptaremos el valor $x$ si se cumple la siguiente desigualdad:

$$
u \leq \frac{f(x)}{Mg(x)}
$$

Se puede verificar que de esta forma se muestrea siguiendo la distribución $p$.

```python
import numpy as np
from matplotlib import pyplot as plt


#Función que evalúa polinomios dados sus coeficientes en orden creciente
def poly(x,cof):
    '''La función regresa la evaluación del número x para el polinomio de grado len(coef)'''
    g = len(cof) #Grado del coeficiente
    y = 0 #Evaluación
    for i in range(g): #Para cada coeficiente 
        y += cof[i]*x**i #Sumamos la evaluación del punto en cada sumando del polinomio
    return y

#Definición del polinomio
x = np.linspace(0,1,50)
coef = [1/12, 1/6, 11/8, 3/2]
y = poly(x, coef)


#Graficación
plt.title("Función de distribución")
plt.xlabel("x"), plt.ylabel("y")
plt.plot(x,y, label = "PDF")
plt.legend()
plt.show()
```

### Ejemplo 1

Proponemos como nuestra función de densidad $p$ al polinomio $f(x)= \frac{1}{12} + \frac{1}{6}x+\frac{11}{8}x^{2}+\frac{3}{2}x^{3}$. Se puede verificar que el soporte es el intervalo $[0,1]$ y que, efectivamente, es una PDF. 

La función de densidad $g$ que utilizaremos para realizar el cómputo es la uniforme en el $[0,1]$. La constante que usaremos para envolver a la función $f$ es $3.125$ (que resulta ser el máximo del polinomio). 

```python
X = np.random.uniform(0.0,1.0, 4500) #Obtenemos una lista de n números siguiendo la distribución g (uniforme)
U = np.random.uniform(0.0, 1.0, 4500) #Obtenemos una lista de n números distribuidos de manera uniforme en [0,1)

cont = 1 #Inicializamos un contador
acep = [] #En una lista guardaremos nuestras aceptaciones o rechazos

while cont <= 4500 and len(acep) < 1000: #Mientras no hayamos recorrido todas las pruebas y mientras no hayamos conseguido 1000 muestras
    t_u = U[cont] #Tomamos un punto en el intervalo de nuestro interés
    t_x = poly(X[cont], coef)/(3.125*1) #Usamos el criterio. Nótese que la constante M es 3.125 y que la evaluación del punto es 1 por ser U([0,1])
    
    if t_u < t_x: #Si pasa el criterio
        acep.append(X[cont]) #Se añade a la lista de aceptación
    cont += 1 #Se suma una aceptación al contador


plt.title("Muestreo de nuestra distribución")
plt.hist(acep)
plt.show()
```

Notamos que, en efecto, la distribución del muestreo sigue a la de nuestra PDF.

### Ejemplo 2

Ahora trabajaremos con la siguiente función de densidad:

$$
f(x):= \frac{3}{4}\left(1-x^{2}\right)\text{    } -1\leq x \leq 1
$$

Se puede verificar que, en efecto, es una función de densidad de probabilidad.

```python
#Nuestra función se ve como sigue

cf2 = [3/4, 0, -3/4] #Coeficientes en orden ascendente
x2 = np.linspace(-1,1,100) #Red para graficar la función
y2 = poly(x2,cf2) #Evaluación de la red


#Gráfica
plt.plot(x2,y2)
plt.title("Segunda PDF"), plt.xlabel("x"), plt.ylabel("3/4(1-x²)")
plt.show()
```

Ahora corremos el algoritmo. De nuevo elegimos la función $g$ como la uniforme, pero ahora en el intervalo [-1,1].

Así, la función es 

$$
g(x)=\frac{1}{2} \text{    } -1\ \leq x \leq 1
$$

Así queda como sigue:

```python
domx = np.linspace(-1,1,1000)
f_x = poly(domx,cf2)
g_x = 1/2*np.ones(1000)
kg_x = 1.5*g_x

plt.plot(domx,f_x, label = "f(x)")
plt.plot(domx,g_x, label = "g(x)")
plt.plot(domx,kg_x, label = "kg(x)")
plt.legend()
plt.title("PDF y función elegida"), plt.xlabel("x"), plt.ylabel("y")
plt.show()
```

```python
X2 = np.random.uniform(-1.0,1.0, 4500) #Obtenemos una lista de n números distribuidos según nuestra $g$
U2 = np.random.uniform(0.0, 1.0, 4500) #Obtenemos una lsita de n números distribuidos de manera uniforme en [0,1)

cont2 = 1 #Inicializamos un contador
acep2 = [] #En una lista guardaremos nuestras aceptaciones o rechazos


while cont2 <= 4500 and len(acep2) < 1000: #Mientras no hayamos recorrido todas las pruebas y mientras no hayamos conseguido 1000 muestras
    t_u = U2[cont2] #Tomamos un punto en el intervalo de nuestro interés
    t_x = poly(X2[cont2], cf2)/(1.5*0.5) #Evaluamos nuestro criterio. En este caso la evaluación de la g es la constante 0.5 y proponemos la constante 1.5
    
    if t_u < t_x: #Si pasa el criterio
        acep2.append(X2[cont2]) #Se añade a la lista de aceptación
    cont2 += 1 #Se suma una aceptación al contador
    
    
    
plt.title("Muestreo")
plt.hist(acep2)
plt.show()
```

Vemos que el muestreo sigue, también, la distribución elegida anteriormente. 

### Conclusiones

Es importante notar que la elección de $g$ depende totalmente de nosotros. Habitualmente será útil elegir una función que sea sencilla de programar o bien, que esté integrada en alguna paquetería de nuestro entorno de programación. Asimismo, la constante $k$ solerá ser aquella para la cuál la distancia entre la función de densidad objetivo y la función de densidad elegida sea tan pequeña como sea posible.

Es importante notar que el uso de este algoritmo está sujeto principalmente a la _suavidad_ o _buen comportamiento_ de la función objetivo dado que en algunos casos puede ocurrir que tengamos picos en la densidad que afecten la eficiencia de nuestro algoritmo.
