Supongamos que tenemos una PDF $p$ de la cuál queremos realizar un muestreo. Por alguna razón, por ejemplo dificultad de cómputo, no nos es posible realizarlo directamente. El método MCMC nos permite realizar este muestreo aprovechando el concepto de cadenas de Markov.


Digamos pues que tenemos la función $f$ (figura 1, en azul) y que deseamos tomar un muestreo a partir de ella. Podríamos seguir el método de aceptación y rechazo y usar, por ejemplo, la función $g$ (en verde) para ello. Sin embargo ocurre que cada paso de nuestro muestreo es indpendiente del anterior y que, siguiendo a $g$, en vez de concentrarnos alrededor de $x=2$, nos concentramos en $x=0$, lo cuál vuelve ineficiente al algoritmo. Más aún, puede ser que la constante que utilicemos para el método de aceptación y rechazo deba ser grande y con ello disminuya aún más la eficiencia del algoritmo.

El algoritmo MCMC busca encontrar la distribución estacionaria de una cadena de Markov que resulte ser precisamente la dada por la función $g$.
